```python
# Importing pandas and matplotlib library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading the dataset from github
df = pd.read_csv("https://raw.githubusercontent.com/shivakrishna67/MS_Assignments/main/diamonds.csv")
```

```python
df
```

##### 1. What proportion of diamonds are between .30 and 1.08 carats?

```python
#Observing first five rows
df.head()
```

```python
# Firstly Iam filtering the filtering the carat between 0.30 and 1.08
df_carats = df[(df['carat'] >= 0.30) & (df['carat'] <= 1.08)]
```

```python
# Secondly Iam finding out the proportion
df_prop = len(df_carats)/len(df)
```

```python
#printing the output
print("Proportion of diamonds between .30 and 1.08: ",df_prop)
```

##### 2. How many of the diamonds have equal x and y dimensions?

```python
#Observing first five rows
df.head()
```

```python
# Filtering the diamonds which are having equal x and y dimensions
df_equal = df[df['x']==df['y']]
```

```python
#Calculating the number of diamonds
no_of_diamonds = len(df_equal)
```

```python
# Printing the output
print("Number of diamonds having equal x and y dimensions:",no_of_diamonds)
```

##### 3. How many of the diamonds have a carat less than the mean carat value?

```python
#Observing first five rows
df.head()
```

```python
# Calculating mean of the carat value
df_carat = df['carat'].mean()
# Filtering the carat which the value is less than mean carat value
df_mean_less = df[df['carat'] < df_carat]
# by counting the number of rows
df_count = len(df_mean_less)
#printing the output
print("Number of diamonds having a carat less than the mean carat value:",df_count)
```

#####  4. How many diamonds have a Premium cut or better? Note that the possible values for the quality of a cut are ranked in ascending order as follows: Fair / Good / Very Good / Premium / Ideal

```python
#Observing first five rows
df.head()
```

```python
# Filter the data for rows with a cut quality of Premium or Ideal
df_cut= df[(df['cut'] == 'Premium') | (df['cut'] == 'Ideal')]

# Count the number of rows in the filtered DataFrame
df_len = len(df_cut)

#Printing the output
print("Number of diamonds having a premium cut or better: ",df_len)
```

##### 5. Which diamond has the highest price per carat? What is its value?

```python
#Observing first five rows
df.head()
```

```python
# Creating a new columns and calculating price per carat
df['price_per_carat'] = df['price'] / df['carat']

# Finding the index of the row with the highest price per carat
index_of_highest_price = df['price_per_carat'].idxmax()

# Retrieve the value of the diamond with the highest price per carat
highest_price_per_carat = df.loc[index_of_highest_price, 'price_per_carat']

print(f"The diamond with the highest price per carat has a value of ${highest_price_per_carat:.2f} per carat.")
```

##### 6.Make a bar plot of the ‘cut’ data. What does it tell us about the frequencies of the individual cut values within the data set?

```python
#Observing first five rows
df.head()
```

```python
# Count the frequencies of individual cut values
cut_counts = df['cut'].value_counts()
```

```python
# Creating a bar plot
cut_counts.plot(kind='bar')
# Customize the plot labels
plt.title('Frequencies of Individual Cut Values')
plt.xlabel('Cut')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

##### 7.  Make boxplots of the diamond price for each cut and discuss any conclusions you can draw from the appearance of the boxplots.

```python
# Creating boxplots for diamond prices grouped by 'cut'
plt.figure(figsize=(10, 6))
ax = df.boxplot(column='price', by='cut', grid=False)

# Customize the plot labels
plt.suptitle('')
plt.title('Boxplots of Diamond Price by Cut')
plt.xlabel('Cut')
plt.ylabel('Price')

# Show the plot
plt.show()
```

**Conclusion** : Boxplot analysis provides insights into the diversity of diamond prices across various cut categories. It enables the identification of both higher and lower median prices, as well as the detection of potential outliers. This examination offers a valuable summary of price distribution within our dataset.

##### 8. Make histograms of the length, width, depth, and total depth attributes and discuss any conclusions you can draw from the shape of the resulting distributions.

```python
df.head()
```

```python
# Create histograms for length, width, depth, and total depth
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Histograms of Diamond Attributes')

df['x'].plot(kind='hist', bins=20, ax=axes[0, 0])
axes[0, 0].set_title('Length (x)')
axes[0, 0].set_xlabel('Length')
axes[0, 0].set_ylabel('Frequency')

df['y'].plot(kind='hist', bins=20, ax=axes[0, 1])
axes[0, 1].set_title('Width (y)')
axes[0, 1].set_xlabel('Width')
axes[0, 1].set_ylabel('Frequency')

df['z'].plot(kind='hist', bins=20, ax=axes[1, 0])
axes[1, 0].set_title('Depth (z)')
axes[1, 0].set_xlabel('Depth')
axes[1, 0].set_ylabel('Frequency')

df['depth'].plot(kind='hist', bins=20, ax=axes[1, 1])
axes[1, 1].set_title('Total Depth (depth)')
axes[1, 1].set_xlabel('Total Depth')
axes[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.subplots_adjust(top=0.9)

# Show the plot
plt.show()
```

**Conclusion**: By providing visual representation Histograms allowing us to central tendency,spread, and potential outliers which helps us to understand the characteristics and make relevant observations from diamond attributes

##### 9.Make a scatter plot of price vs. carat. What can we say about the relationship between those two attributes?

```python
# Creating a scatter plot of price vs. carat
plt.figure(figsize=(10, 6))
plt.scatter(df['carat'], df['price'], alpha=0.5)
plt.title('Scatter Plot of Price vs. Carat')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.grid(True)

# Showing the plot
plt.show()
```

**Conclusion**: The scatter plot indicates a positive correlation between carat and price, yet there's noticeable price variation for diamonds with similar carat weights, suggesting the influence of other factors on diamond pricing.
